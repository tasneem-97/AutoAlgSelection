```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import time
import pandas as pd 
stage1b_start = time.perf_counter()
```

```python
import os
from csv import DictWriter

#MyList=['iris.csv','breast_cancer.csv','heart.csv','diabetes.csv']
#MyList=['zoo.csv','yeast.csv','wine.csv','hapetitis.csv','ionospere.csv','indian liver.csv','horse-ecolic surgical.csv','hayes-roth.csv','glass.csv','flags.csv','fertility.csv','cmc.csv','bridges1.csv','bridges2.csv','braest-cancer wincosin.csv','autos.csv','austrlian statelog.csv','audiology.csv','Annealing.csv','Acute-urinary.csv','acute-inflammation.csv','abalone.csv','iris.csv','breast_cancer.csv','heart.csv',"diabetes.csv","chess.csv","statlog-heart.csv","German Credit Data.csv","transfusion.csv","car_evaluation.csv","BankNote_Authentication.csv","balance-scaledata.csv",'Absenteeism_at_work.csv','Algerian_forest_fires_dataset_UPDATE.csv','Autism-Adult-Data.csv','ballons.csv','dataR2.csv','sobar-72.csv','risk_factors_cervical_cancer.csv','cirrhosis.csv','pop_failure.csv','house_voting.csv','cylinder_bands.csv','dermatology.csv','Thyroid_Diff.csv','diabetes_data_upload.csv','ecoli.csv','ObesityDataSet_raw_and_data_sinthetic.csv','haberman.csv']

import glob

folder_path = "balancedData"
csv_files = glob.glob(folder_path + "/*.csv")  # Adjust for different file types
file_names = [os.path.basename(file) for file in csv_files]

print(csv_files[0])

mydata=0
outputKNNList=[]
outputGNBList=[]
outputSVCList=[]
outputlRList=[]
outputDTList=[]
outputGBCList=[]
outputRFList=[]
AccDataset={}
algAccOut={}
for i in range(len(file_names)):
    mydata=pd.read_csv("balancedData/"+file_names[i])
    #splitting data
    x= mydata.iloc[:,1:-1].to_numpy()
    y=mydata.iloc[:,-1].to_numpy()
    #encoding data 
    
    # Create KNN classifier
    knn = KNeighborsClassifier(n_neighbors = 5)
    KNNscores=cross_val_score(knn,x,y,cv=3,scoring='accuracy' )
    
    #printing avg score
    avgKNNscores=round(np.mean(KNNscores),2)
    outputKNNList.append(avgKNNscores)
   # print("Knn Avg scores",avgKNNscores)
    #Create GaussianNB classifier
    gnb = GaussianNB()
    gnbscores=cross_val_score(gnb,x,y,cv=3,scoring='accuracy' )
    #print("GNB scores",gnbscores)
    avgGNBscores=round(np.mean(gnbscores),2)
    outputGNBList.append(avgGNBscores)
    #print("GNB Avg scores",avgGNBscores)
    #create SVC(support vector classifier
    SVCscores=cross_val_score(SVC(),x,y,cv=3,scoring='accuracy')
    avgSVCscores=round(np.mean(SVCscores),2)
    outputSVCList.append(avgSVCscores)
    
    #logistic Regresion
    # instantiate & fit
    lr=LogisticRegression(max_iter=1000)
    LrScore=cross_val_score(lr,x,y,cv=3,scoring='accuracy' )
    avglrScore=round(np.mean(LrScore),2)
    #print('lrscore',avglrScore)
    outputlRList.append(avglrScore)
    
    #Desion tree
    DT = DecisionTreeClassifier(min_samples_split=10,max_depth=3)
    DTScore=cross_val_score(DT,x,y,cv=3, scoring='accuracy')
    avgDTScore=round(np.mean(DTScore),2)
    #print('avgDtscore',avgDTScore)
    outputDTList.append(avgDTScore)

    #Random forest
# instantiate & fit
    RF = RandomForestClassifier(n_estimators=300,max_depth=3)
    RFScore=cross_val_score(RF,x,y,cv=3 ,scoring='accuracy')
    avgRFScore=round(np.mean(RFScore),2)
    #print('avgRFScore',avgRFScore)
    outputRFList.append(avgRFScore)
    
    #Gradient boosting
    gbc = GradientBoostingClassifier(n_estimators=100)
    GBCScore=cross_val_score(gbc,x,y,cv=3 ,scoring='accuracy')
    avgGBCScore=round(np.mean(GBCScore),2)
    #print('avgGBCScore',avgGBCScore)
    outputGBCList.append(avgGBCScore)
    
print(outputlRList[0])  
print(outputlRList[1])
print(outputlRList[2])
```

```python
AccDataset = {}
#outer_keys = ['first', 'second', 'third']
inner_keys = ['KNN', 'GNB', 'SVC', 'LR', 'DT', 'RF','GBC']
values = [outputKNNList,outputGNBList,outputSVCList,outputlRList,outputDTList,outputRFList,outputGBCList]

# Using a for loop to add inner dictionaries to the outer dictionary
j=0
while j<11:
    for outer_key in file_names:
        AccDataset[outer_key] = {}  # Create an empty inner dictionary
        for i in range(len(inner_keys)):  
            #print(values[i][j])
            AccDataset[outer_key][inner_keys[i]] = values[i][j]
        j+=1
print(AccDataset)
```

```python
#['LR', 'SVC', 'GBC', 'GBC', 'GBC', 'GNB', 'LR', 'LR', 'KNN', 'SVC', 'KNN', 'RF']
maxAlgNames=[]
maxAlgAcc=[]
for dataName,algName in AccDataset.items():
    #print(dataName,algName)
    for nameMaxalg in AccDataset[dataName]:
         #print(max(AccDataset[dataName].values()))
         if AccDataset[dataName][nameMaxalg] == max(AccDataset[dataName].values()): 
             maxAlgNames.append(nameMaxalg)
             maxAlgAcc.append(max(AccDataset[dataName].values()))
             break
            
         
           
            
print(maxAlgNames)
print(maxAlgAcc)


    #saving  output in file
data = {
    'DatasetName': file_names,
    'KNN': outputKNNList,
    'GNB':outputGNBList,
    'SVC':outputSVCList,
    'LR':outputlRList,
    'DT':outputDTList,
    'RF':outputRFList,
    'GBC':outputGBCList, 
    'max Acc':maxAlgAcc,
    'maxAlgNames':maxAlgNames
}
out_data=pd.DataFrame(data)
print(out_data)
file_name = 'outputs/FirstStageOutData.csv'
out_data.to_csv(file_name)
print('DataFrame is written to Excel File successfully.')

stage1b_end = time.perf_counter()
stage1b_time = stage1b_end - stage1b_start

print("Stage 1B - Base Algorithms Evaluation Time:", stage1b_time, "seconds")

timing_df = pd.DataFrame([{
    "Stage": "Stage 1B - Base Algorithms Evaluation",
    "Execution_Time_sec": stage1b_time
}])

timing_df.to_csv("stage1b_time.csv", index=False)
```
