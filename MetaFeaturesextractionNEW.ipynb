```python
import os
import glob
import pandas as pd
import numpy as np


# -------- STEP 1: تقسيم الملفات الكبيرة إن وجدت -------- #
folder_path = "balancedData"
output_dir = os.path.join(folder_path, "split_files")
os.makedirs(output_dir, exist_ok=True)

max_rows = 15000
chunk_size = 10000

csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
complexity_start = time.perf_counter()
for file in csv_files:
    df = pd.read_csv(file)
    if len(df) > max_rows:
        base_name = os.path.splitext(os.path.basename(file))[0]
        for i in range(0, len(df), chunk_size):
            chunk = df.iloc[i:i+chunk_size]
            chunk_file = os.path.join(output_dir, f"{base_name}_part_{i//chunk_size + 1}.csv")
            chunk.to_csv(chunk_file, index=False)
    else:
        new_path = os.path.join(output_dir, os.path.basename(file))
        df.to_csv(new_path, index=False)

print("✅ تم تجهيز الملفات الصغيرة في:", output_dir)

# -------- STEP 2: استخراج meta-features -------- #
folder_path = output_dir
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
file_names = [os.path.basename(file) for file in csv_files]

measureNamelist = []
all_dataset_values = {}

for file in file_names:
    mydata = pd.read_csv(os.path.join(folder_path, file))

    if len(mydata) > 1000:
        mydata = mydata.sample(n=1000, random_state=42)

    x = mydata.iloc[:, :-1].to_numpy()
    y = mydata.iloc[:, -1].to_numpy()
    selected_features = ["f1", "n2",  "linsep", "t1"]

    mfe = MFE(groups="complexity", features=selected_features, summary=["nanmean"])
    mfe.fit(x, y)
    ft = mfe.extract(n_jobs=1)

    if not measureNamelist:
        measureNamelist = ft[0]

    # خزّن القيم لكل ملف
    values = []
    for val in ft[1]:
        values.append(round(val, 2) if not np.isnan(val) else np.nan)
    
    all_dataset_values[file] = values

# -------- STEP 3: بناء الـ DataFrame النهائي -------- #
output_dict = {"measureName": measureNamelist}

for dataset_name, values in all_dataset_values.items():
    output_dict[dataset_name] = values

out_data = pd.DataFrame(output_dict)
# عرض الصفوف اللي فيها أي NaN


#cleaned=out_data.dropna()
# -------- STEP 4: حفظ النتائج -------- #
os.makedirs("outputs/MFEOut", exist_ok=True)
file_name = 'outputs/MFEOut/complexity.csv'
out_data.to_csv(file_name, index=False)

print("📁 تم حفظ ملف meta-features بالشكل المطلوب في:", file_name)
print("📊 الشكل النهائي:", out_data.shape)
```

```python
from sklearn.impute import SimpleImputer
import pandas as pd

# قراءة البيانات
data = pd.read_csv('outputs/MFEOut/complexity.csv')

# فصل العمود الأول لأنه يحتوي على أسماء المقاييس (strings)
measure_names = data['measureName']
features_only = data.drop(columns=['measureName'])

# إنشاء imputer لتعبئة القيم المفقودة بالمتوسط
imputer = SimpleImputer(strategy='mean')

# تطبيق التعويض
imputed_data = imputer.fit_transform(features_only)

# تحويل النتائج مرة أخرى إلى DataFrame
filled_df = pd.DataFrame(imputed_data, columns=features_only.columns)

# إرجاع عمود measureName
filled_df.insert(0, 'measureName', measure_names)

# حفظ النتائج
filled_df.to_csv('outputs/MFEOut/complexity_filled.csv', index=False)
print("✅ تم تعويض القيم المفقودة وحفظ الملف الجديد.")
```

Complexity MetaFeatures


```python
import pandas as pd
import os

# ----- STEP 1: تحميل ملف meta-features -----
meta_features_path = "outputs/MFEOut/complexity_filled.csv"
meta_df = pd.read_csv(meta_features_path)

# ----- STEP 2: تحويل الملف بحيث كل صف = ملف، وكل عمود = مقياس -----
meta_df = meta_df.set_index("measureName").T
meta_df.index.name = "split_file_name"
meta_df.reset_index(inplace=True)

# ----- STEP 3: استخراج الاسم الأصلي للداتاسيت من اسم الملف المقسم -----
# مثلاً: "iris_part_1.csv" --> "iris"
meta_df["original_dataset"] = meta_df["split_file_name"].apply(lambda x: x.split("_part_")[0] if "_part_" in x else x.split(".csv")[0])

# ----- STEP 4: حساب المتوسط لكل داتاسيت أصلي -----
# نستبعد عمود اسم الملف والاسم الأصلي ونحسب المتوسط للباقي
features_only = meta_df.drop(columns=["split_file_name", "original_dataset"])
meta_df_avg = meta_df.groupby("original_dataset")[features_only.columns].mean().reset_index()

# ----- STEP 5: حفظ الملف النهائي -----
os.makedirs("outputs/MFEOut", exist_ok=True)
output_file = "outputs/MFEOut/complexity_avg_by_dataset.csv"
meta_df_avg.to_csv(output_file, index=False)

print("✅ تم حفظ المتوسطات النهائية لكل داتاسيت في:", output_file)
print("📊 الشكل النهائي:", meta_df_avg.shape)
complexity_end = time.perf_counter()
timing_rows.append({"Stage": "Complexity Meta-Features", "Execution_Time_sec": complexity_end - complexity_start})

# =========================
# Total Stage 2
# =========================
stage2_end = time.perf_counter()
timing_rows.append({"Stage": "Stage 2 Total", "Execution_Time_sec": stage2_end - stage2_start})

timing_df = pd.DataFrame(timing_rows)
print(timing_df)
timing_df.to_csv("stage2_time.csv", index=False)
```
