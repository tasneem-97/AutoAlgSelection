```python
import pandas as pd

df = pd.read_csv("outputs/FirstStageOutData.csv")

df = (
    df.drop_duplicates()
      .dropna(how="all")
      .reset_index(drop=True)
)

models = ["KNN", "GNB", "SVC", "LR", "DT", "RF", "GBC"]

exclusive_wins = {model: 0 for model in models}
ties_with_best = {model: 0 for model in models}

for _, row in df.iterrows():

    accuracies = {model: row[model] for model in models}

    max_acc = max(accuracies.values())

    best_models = [m for m, acc in accuracies.items() if acc == max_acc]

    if len(best_models) == 1:
        exclusive_wins[best_models[0]] += 1
    else:
        for model in best_models:
            ties_with_best[model] += 1

result_df = pd.DataFrame({
    "Classifier": models,
    "Exclusive Wins": [exclusive_wins[m] for m in models],
    "Ties with the Best": [ties_with_best[m] for m in models]
})

# صف الإجمالي
total_row = pd.DataFrame([{
    "Classifier": "Total",
    "Exclusive Wins": sum(exclusive_wins.values()),
    "Ties with the Best": sum(1 for _, row in df.iterrows()
                              if len([m for m in models
                                      if row[m] == max(row[models])]) > 1)
}])

result_df = pd.concat([result_df, total_row], ignore_index=True)

result_df.to_csv(
    "classifier_wins_and_ties.csv",
    index=False
)

print(result_df)
```
